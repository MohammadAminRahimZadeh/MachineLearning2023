Problem 10: Given a Pandas DataFrame containing information about students, add a new column 'Grade' based on the following criteria:

If 'Score' >= 90, Grade is 'A'
If 'Score' >= 80, Grade is 'B'
If 'Score' >= 70, Grade is 'C'
If 'Score' < 70, Grade is 'D'

```python
data = {
    'Name': ["Aria","reza","ahmad","morteza", "abbas"],
    'number': [92, 88, 60, 75, 85]
}
df= pd.DataFrame(data)
print(df)
print("-------------------------")
df['Grade'] = df['number'].apply(lambda score: 'A' if score >= 90 else ('B' if score >= 80 else ('C' if score >= 70 else 'D')))
print(df)
```

Problem 11: You are given a dataset containing monthly sales data for three products (A, B, and C) over a two-year period. Your task is to perform various data analysis tasks using NumPy, Pandas, and Matplotlib.

Dataset:
import pandas as pd

data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)
Tasks:
Calculate and display the total sales for each product over the two-year period.

Calculate and display the average monthly sales for each product.

Find the month with the highest sales for each product, and display the product and the sales value for that month.

Calculate and display the percentage change in sales for each product from January to December in the second year (2022).

Create a line chart using Matplotlib to visualize the monthly sales data for each product over the two-year period. Label the axes and provide a title for the chart.

Calculate the correlation between Product A and Product B sales. Is there a strong correlation between these two products? Display the correlation coefficient.

```python
data = {
    'Month': pd.date_range(start='2021-01-01', periods=24, freq='M'),
    'Product A Sales': [500, 480, 600, 750, 900, 850, 920, 1100, 1300, 1350, 1500, 1450, 1550, 1600, 1650, 1600, 1500, 1400, 1600, 1700, 1800, 1750, 1850, 1900],
    'Product B Sales': [300, 320, 400, 450, 500, 580, 700, 750, 820, 900, 950, 980, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1500, 1600, 1550, 1700],
    'Product C Sales': [200, 210, 250, 280, 320, 350, 380, 400, 420, 440, 460, 480, 500, 520, 540, 560, 580, 600, 620, 640, 660, 680, 700, 720]
}

sales_df = pd.DataFrame(data)

#Calculate and display the average monthly sales for each product
print("total sales each product:\n",  sales_df[['Product A Sales', 'Product B Sales', 'Product C Sales']].sum())
print("\naverage sales for each product:\n", sales_df[['Product A Sales', 'Product B Sales', 'Product C Sales']].mean())
#Find the month with the highest sales for each product, and display the product and the sales value for that month
idx_product_a_max = sales_df['Product A Sales'].idxmax()
idx_product_b_max = sales_df['Product B Sales'].idxmax()
idx_product_c_max = sales_df['Product C Sales'].idxmax()
print("\n month with highest sale")
print(" Product A:", sales_df.loc[idx_product_a_max, 'Month'].strftime('%Y-%m-%d'), "with sales of", sales_df.loc[idx_product_a_max, 'Product A Sales'])
print(" Product B:", sales_df.loc[idx_product_b_max, 'Month'].strftime('%Y-%m-%d'), "with sales of", sales_df.loc[idx_product_b_max, 'Product B Sales'])
print(" Product C:", sales_df.loc[idx_product_c_max, 'Month'].strftime('%Y-%m-%d'), "with sales of", sales_df.loc[idx_product_c_max, 'Product C Sales'])
#Calculate and display the percentage change in sales for each product from January to December in the second year (2022)
jan_2022 = sales_df[sales_df['Month'] == '2022-01-31']
dec_2022 = sales_df[sales_df['Month'] == '2022-12-31']
percentage_change = ((dec_2022[['Product A Sales', 'Product B Sales', 'Product C Sales']].values -
                     jan_2022[['Product A Sales', 'Product B Sales', 'Product C Sales']].values) /
                     jan_2022[['Product A Sales', 'Product B Sales', 'Product C Sales']].values) * 100
print("\n Percentage Change in Sales for each product: January to December in 2022:")
print(pd.DataFrame(percentage_change, columns=['Product A Sales', 'Product B Sales', 'Product C Sales']))
#Create a line chart using Matplotlib to visualize the monthly sales data for each product over the two-year period. Label the axes and provide a title for the chart
plt.figure(figsize=(10, 6))
plt.plot(sales_df['Month'], sales_df['Product A Sales'], label='Product A Sales', marker='o')
plt.plot(sales_df['Month'], sales_df['Product B Sales'], label='Product B Sales', marker='s')
plt.plot(sales_df['Month'], sales_df['Product C Sales'], label='Product C Sales', marker='d')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Data for Each Product')
plt.legend()
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%b %Y'))
plt.tight_layout()
plt.show()
#Calculate the correlation between Product A and Product B sales. Is there a strong correlation between these two products? Display the correlation coefficient
correlation = sales_df['Product A Sales'].corr(sales_df['Product B Sales'])
print("\ncorrelation between Product A and Product B Sales is:", correlation)




```
