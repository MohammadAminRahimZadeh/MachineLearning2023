```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.datasets import make_classification
import sklearn.metrics as met
from sklearn.metrics import confusion_matrix
from mlxtend.plotting import plot_decision_regions
```

**Part1**

```python

X, y = make_classification(n_samples=1000, n_features=2, n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, class_sep=1,random_state=3)
```

```python
colors = np.array(['green','orange'])
plt.scatter(X[:,0],X[:,1],c=colors[y])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.2 , random_state=3 )
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

**part2**

```python
model1 = LogisticRegression(max_iter=1000, random_state =3)
model1.fit(x_train,y_train);
y_pred = model1.predict(x_test)
```

```python
print("iterations:", model1.n_iter_)
print("train score:", model1.score(x_train,y_train))
print("test score:", model1.score(x_test,y_test))
print(" Mean squared error:",met.mean_squared_error(y_test, y_pred))
print("confusion_matrix:",confusion_matrix(y_test, y_pred))
```

```python
model2 = SGDClassifier(loss='log_loss', max_iter=1000, eta0=0.1, random_state =3)
model2.fit(x_train,y_train);
y2_pred = model2.predict(x_test)
```

```python
print("iterations:", model2.n_iter_)
print("train score:", model2.score(x_train,y_train))
print("test score:", model2.score(x_test,y_test))
print(" Mean squared error:",met.mean_squared_error(y_test, y2_pred))
print("confusion_matrix:",confusion_matrix(y_test, y2_pred))
```

**Part3**


```python
x1_min, x2_min = x_test.min(0)
x1_max, x2_max = x_test.max(0)
n=500
x1r = np.linspace(x1_min, x1_max, n)
x2r = np.linspace(x2_min, x2_max, n)
x1m , x2m = np.meshgrid(x1r, x2r)
xm = np.stack((x1m.flatten(), x2m.flatten()), axis=1)
ym = model.decision_function(xm)
colors = np.array(['green','orange'])
plt.scatter(x_test[:,0],x_test[:,1], c=colors[y_test])
plt.contour(x1m, x2m, ym.reshape(x1m.shape), levels=[-1,0,1])
plt.show
```

```python
y_pred1 = model.predict(X)
misclassified1 = y_pred1 != y
plot_decision_regions(X, y, clf=model, markers='xo', X_highlight=X[misclassified1])
plt.scatter(X[misclassified1][:, 0], X[misclassified1][:, 1], c='red', marker='o', edgecolor='black', s=30, label='Misclassified')
plt.show()
```
